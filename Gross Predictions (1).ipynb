```python
X=df.loc[:,df.columns!="Gross"]
Y=df["Gross"]
```

```python
import statsmodels.formula.api as sm     #BACKWARD ELMINATION
X=np.append(arr=np.ones((3743,1)).astype(float),values=X,axis=1)
XT_opt=X[:,:].astype(float)
Y=Y.astype(float)
reg_OLS=sm.OLS(endog=Y,exog=XT_opt).fit()
```

```python
reg_OLS.summary()
```

```python
import statsmodels.formula.api as sm
def backwardElimination(x, sl):
    numVars = len(x[0])
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(Y, x).fit()
        maxVar = max(regressor_OLS.pvalues)
        if maxVar > sl:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1)
    print(regressor_OLS.summary())
    return x
```

```python
SL = 0.05
XT_opt=X[:,:]
X_Modeled = backwardElimination(XT_opt, SL)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_Modeled,Y,test_size=0.2,random_state=0)
```

```python

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
Y_pred=reg.predict(X_test)
mean_squared_error(Y_pred,Y_test)
```
